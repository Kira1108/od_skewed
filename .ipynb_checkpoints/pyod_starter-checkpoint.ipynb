```python
import numpy as np
import pandas as pd
import pyod
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
%config InlineBackend.figure_format = 'retina'
from pyod.models.ecod import ECOD
from sklearn.metrics import roc_auc_score, accuracy_score, confusion_matrix
```

```python
def plot_scatter(X, y, title = 'Dataset Distribution'):
    sns.scatterplot(x = X[:,0],y = X[:,1],hue = y,s = 100, alpha = 0.9)
    plt.grid(linestyle = '--')
    plt.title(title)
```

```python
X,y = make_blobs(
    n_samples = 300,
    n_features = 2,
    centers = [[3,0],[5,2]])

X_anom = np.random.uniform(X.min()-1,X.max() + 1, size= (60,2))
y_anom = np.array([y.max() + 1] * len(X_anom))

X = np.r_[X, X_anom]
y = np.r_[y, y_anom]
y = (y == 2).astype(int)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.4)
```

```python
clf = ECOD(contamination=0.2)
clf.fit(X_train)
pred = clf.predict(X_test)
scores = clf.decision_function(X_test)
```

```python
fig, ax = plt.subplots(figsize= (12,10))
plt.subplot(2,2,1)
plot_scatter(X, y,'Raw Dataset')
plt.subplot(2,2,2)
plot_scatter(X_test, y_test,'Test set ground truth')
plt.subplot(2,2,3)
plot_scatter(X_test, scores,'Test set score')
plt.subplot(2,2,4)
plot_scatter(X_test, pred,'Test set prediction')
```

### Evaluation on right skewed dataset

```python
def right_skew(
    mu = 10,
    sigma = 1.5,
    n_normal = 100,
    n_anom = 20,
    low_sigma = 3,
    high_sigma = 10):

    x = np.random.normal(mu,sigma,n_normal)
    anom = np.random.uniform(mu  + low_sigma * sigma, mu + high_sigma * sigma,n_anom)
    x = np.r_[x, anom]
    return x,np.array([0]* n_normal + [1] * n_anom)
    
x,y = right_skew()  
sns.kdeplot(x, lw = 3)
plt.show()
```

```python
x1,y1 = right_skew(100,10)
x2,y2 = right_skew(200,10)
x3,y3 = right_skew(300,10)
x = np.c_[x1, x2, x3]
y = np.c_[y1, y2, y3]
y = (y.sum(axis = -1) > 1).astype(int)
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.4)
```

### ECOD performance -  good enough

```python
clf = ECOD(contamination=0.2)
clf.fit(X_train)

acc  = accuracy_score(y_test, clf.predict(X_test))
auc = roc_auc_score(y_test, clf.predict(X_test))

print("Accuracy = {}".format(acc))
print("ROC AUC = {}".format(auc))

sns.heatmap(confusion_matrix(y_test, clf.predict(X_test)), annot=True)
plt.show()
```

### Isolation forest performs best

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(contamination=0.2)
iso.fit(X_train)
pred = iso.predict(X_test)
pred = np.array([1 if v == -1 else 0 for v in pred])

acc  = accuracy_score(y_test, pred)
auc = roc_auc_score(y_test, pred)
print("Accuracy = {}".format(acc))
print("ROC AUC = {}".format(auc))
sns.heatmap(confusion_matrix(y_test, pred), annot=True)
plt.show()
```

### OneClass SVM dose a relatively worse job on skewed dataset

```python
from sklearn.svm import OneClassSVM
```

```python
svm = OneClassSVM()

svm.fit(X_train)

pred = svm.predict(X_test)
pred = np.array([1 if v == -1 else 0 for v in pred])
acc  = accuracy_score(y_test, pred)
auc = roc_auc_score(y_test, pred)
print("Accuracy = {}".format(acc))
print("ROC AUC = {}".format(auc))
sns.heatmap(confusion_matrix(y_test, pred), annot=True)
plt.show()
```
